```python
# ====================================================================
# 0. å‰æã¨ãªã‚‹å®šç¾©ã¨é–¢æ•°
# ====================================================================
import bpy
import math
import numpy as np
from mathutils import Vector, Quaternion

# ã€é‡è¦ã€‘ã‚­ãƒ¼ãƒã‚¤ãƒ³ãƒˆã‚¤ãƒ³ãƒ‡ãƒƒã‚¯ã‚¹ã¨ãƒœãƒ¼ãƒ³åã®å¯¾å¿œä»˜ã‘
# keypoints_listã®ã‚¤ãƒ³ãƒ‡ãƒƒã‚¯ã‚¹ (0-16) ã¨ã€ã‚¢ãƒ¼ãƒãƒãƒ¥ã‚¢ã®ãƒœãƒ¼ãƒ³å (Tail) ã®ãƒãƒƒãƒ”ãƒ³ã‚°
# ã“ã®ãƒãƒƒãƒ”ãƒ³ã‚°ã¯ã€ã”ä½¿ç”¨ã®ãƒ¢ãƒ‡ãƒ« (R6ãªã©) ã«åˆã‚ã›ã¦æ­£ç¢ºã«èª¿æ•´ã—ã¦ãã ã•ã„ã€‚
BONE_INDEX_MAP = {
    # 0: 'Hip' ã¯ãƒ«ãƒ¼ãƒˆä½ç½®ã«ä½¿ã†ã“ã¨ãŒå¤šã„
    'spine.001': 7, 
    'spine.002': 8,
    'head.001': 9,
    'head.002': 10,
    'waist.001.l': 4, # Hipsã®ä½ç½®ã‹ã‚‰å›è»¢ã‚’è¨ˆç®—ã—ãªã„ãŸã‚ã€ã“ã“ã§ã¯0ã‚’ä½¿ç”¨
    'waist.001.r': 1, # Hipsã®ä½ç½®ã‹ã‚‰å›è»¢ã‚’è¨ˆç®—ã—ãªã„ãŸã‚ã€ã“ã“ã§ã¯0ã‚’ä½¿ç”¨
    
    # å·¦è…• (ä¾‹)
    'shoulder.001.l': 11, # L Shoulder
    'arm.001.l': 12,   # L Elbow
    'arm.002.l': 13,      # L Wrist
    
    # å³è…• (ä¾‹)
    'shoulder.001.r': 14, # L Shoulder
    'arm.001.r': 15,   # L Elbow
    'arm.002.r': 16,      # L Wrist
    
    # å·¦è„š (ä¾‹)
    'leg.001.l': 5, # L Hip (ã‚­ãƒ¼ãƒã‚¤ãƒ³ãƒˆ)
    'leg.002.l': 6,# L Knee
    'feet.001.l': 0,  # L Ankle
    
    # å·¦è„š (ä¾‹)
    'leg.001.r': 2, # L Hip (ã‚­ãƒ¼ãƒã‚¤ãƒ³ãƒˆ)
    'leg.002.r': 3,  # L Knee
    'feet.001.r': 0,  # L Ankle
}

import numpy as np
import bpy # Blenderã®ãƒ¢ã‚¸ãƒ¥ãƒ¼ãƒ«ãŒå¿…è¦ãªãŸã‚ã€ã‚³ãƒ¼ãƒ‰å…¨ä½“ã§ã‚¤ãƒ³ãƒãƒ¼ãƒˆãŒå¿…è¦ã§ã™
from mathutils import Vector

def load_keypoints_data_from_npz(filepath, scale_factor=1):
    """NPZãƒ•ã‚¡ã‚¤ãƒ«ã‹ã‚‰ã‚­ãƒ¼ãƒã‚¤ãƒ³ãƒˆãƒ‡ãƒ¼ã‚¿ã‚’èª­ã¿è¾¼ã¿ã€ã‚¹ã‚±ãƒ¼ãƒ«èª¿æ•´ã—ã€Vectorã®ãƒªã‚¹ãƒˆã«å¤‰æ›ã™ã‚‹"""
    try:
        data = np.load(filepath)
        keypoints_data = None
        
        # ã©ã®ã‚­ãƒ¼åã§ãƒ‡ãƒ¼ã‚¿ãŒä¿å­˜ã•ã‚Œã¦ã„ã‚‹ã‹ã‚’ç¢ºèª
        if 'keypoints_4d' in data:
            # 4æ¬¡å…ƒãƒ‡ãƒ¼ã‚¿ (N, 17, 4) ã®å ´åˆ
            keypoints_data = data['keypoints_4d']
        elif 'keypoints_3d' in data:
            # 3æ¬¡å…ƒãƒ‡ãƒ¼ã‚¿ (N, 17, 3) ã®å ´åˆ
            keypoints_data = data['keypoints_3d']
        else:
            print(f"ã‚¨ãƒ©ãƒ¼: NPZãƒ•ã‚¡ã‚¤ãƒ«ã« 'keypoints_4d' ã¾ãŸã¯ 'keypoints_3d' ã‚­ãƒ¼ãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“ã€‚")
            return None

        # è¤‡æ•°ã®ãƒ•ãƒ¬ãƒ¼ãƒ ãŒå«ã¾ã‚Œã¦ã„ã‚‹å ´åˆã€æœ€åˆã®ãƒ•ãƒ¬ãƒ¼ãƒ  ([0]) ã‚’å–å¾—
        if keypoints_data.ndim == 3:
            keypoints_data = keypoints_data[0] 
        
        # ã€æ©Ÿèƒ½è¿½åŠ ã€‘: 4åˆ—ç›®ï¼ˆä¿¡é ¼åº¦ã‚¹ã‚³ã‚¢ï¼‰ã‚’å‰Šé™¤ã™ã‚‹å‡¦ç†
        if keypoints_data.shape[-1] == 4:
            print("ğŸ’¡ ãƒ‡ãƒ¼ã‚¿ã«4åˆ—ç›®ï¼ˆä¿¡é ¼åº¦ï¼‰ãŒå«ã¾ã‚Œã¦ã„ã¾ã™ã€‚3æ¬¡å…ƒåº§æ¨™ã®ã¿ã«æ•´å½¢ã—ã¾ã™ã€‚")
            # æœ€å¾Œã®è»¸ã®æœ€åˆã®3åˆ— (X, Y, Z) ã®ã¿ã‚’é¸æŠ
            keypoints_data = keypoints_data[:, :3] 

        # æœ€çµ‚çš„ãªå½¢çŠ¶ã®æ¤œè¨¼
        if keypoints_data.shape != (17, 3):
            print(f"ã‚¨ãƒ©ãƒ¼: å‡¦ç†å¾Œã®ã‚­ãƒ¼ãƒã‚¤ãƒ³ãƒˆé…åˆ—ã®å½¢çŠ¶ãŒ (17, 3) ã§ã¯ã‚ã‚Šã¾ã›ã‚“: {keypoints_data.shape}")
            return None
            
        # åº§æ¨™ã®ã‚¹ã‚±ãƒ¼ãƒ«èª¿æ•´
        keypoints_array_scaled = keypoints_data * scale_factor
            
        # NumPyé…åˆ—ã‚’ mathutils.Vector ã®ãƒªã‚¹ãƒˆã«å¤‰æ›
        keypoints_list = [Vector(kp) for kp in keypoints_array_scaled]
        
        print(f"âœ… ã‚­ãƒ¼ãƒã‚¤ãƒ³ãƒˆã‚’ã‚¹ã‚±ãƒ¼ãƒ«ãƒ•ã‚¡ã‚¯ã‚¿ {scale_factor} ã§ãƒ­ãƒ¼ãƒ‰ã—ã¾ã—ãŸã€‚")
        return keypoints_list
        
    except Exception as e:
        print(f"âŒ ã‚¨ãƒ©ãƒ¼: NPZãƒ•ã‚¡ã‚¤ãƒ«ã®èª­ã¿è¾¼ã¿ã«å¤±æ•—ã—ã¾ã—ãŸ: {e}")
        return None

# --- UIæ“ä½œã®ã‚³ãƒ¼ãƒ‰ï¼ˆå¿…è¦ã§ã‚ã‚Œã°ï¼‰ ---
# bpy.ops.wm.call_panel(name="VIEW3D_PT_gizmo_display")

#def calculate_rotation(pbone, target_tail_world_loc):
#    """
#    ç¾åœ¨ã®ãƒœãƒ¼ãƒ³ã®è¦ªã®ãƒˆãƒ©ãƒ³ã‚¹ãƒ•ã‚©ãƒ¼ãƒ ã‚’è€ƒæ…®ã—ã€ãƒ¯ãƒ¼ãƒ«ãƒ‰åº§æ¨™ã®ç›®æ¨™ç‚¹ã«å‘ã‘ã¦ãƒœãƒ¼ãƒ³ã‚’å›è»¢ã•ã›ã‚‹ã‚¯ã‚©ãƒ¼ã‚¿ãƒ‹ã‚ªãƒ³ã‚’è¨ˆç®—ã™ã‚‹
#    """
#    # ç¾åœ¨ã®ãƒœãƒ¼ãƒ³ã®ãƒ¯ãƒ¼ãƒ«ãƒ‰åº§æ¨™ã®ãƒ™ã‚¯ãƒˆãƒ« (ãƒœãƒ¼ãƒ³ã®æ–¹å‘)
#    current_vec = pbone.tail_cache - pbone.head_cache # World space bone vector
#    
#    # ç›®æ¨™ã¨ãªã‚‹ãƒ¯ãƒ¼ãƒ«ãƒ‰åº§æ¨™ã®ãƒ™ã‚¯ãƒˆãƒ«
#    # ç›®æ¨™ç‚¹ - ãƒœãƒ¼ãƒ³ã®ãƒ˜ãƒƒãƒ‰ã®ãƒ¯ãƒ¼ãƒ«ãƒ‰åº§æ¨™ (ãƒ˜ãƒƒãƒ‰ã®ä½ç½®ã¯è¦ªã®å›è»¢ã«ã‚ˆã‚Šã™ã§ã«æ›´æ–°ã•ã‚Œã¦ã„ã‚‹)
#    target_vec = target_tail_world_loc - pbone.head_cache
#    
#    # 2ã¤ã®ãƒ™ã‚¯ãƒˆãƒ«é–“ã®å›è»¢ã‚’è¨ˆç®—
#    # rotation_differenceã¯ã€current_vecã‚’target_vecã«å›è»¢ã•ã›ã‚‹ã‚¯ã‚©ãƒ¼ã‚¿ãƒ‹ã‚ªãƒ³ã‚’è¿”ã™
#    local_quat = current_vec.rotation_difference(target_vec)
#    
#    # å›è»¢ã‚’ãƒœãƒ¼ãƒ³ã®ãƒ­ãƒ¼ã‚«ãƒ«ç©ºé–“ã«é©ç”¨ã™ã‚‹ãŸã‚ã«ã€è¦ªã®é€†å›è»¢ã‚’é©ç”¨
#    if pbone.parent:
#        # è¦ªã®ãƒ¯ãƒ¼ãƒ«ãƒ‰å›è»¢ã®é€†è¡Œåˆ—ã‚’ä½¿ã£ã¦ã€ãƒ­ãƒ¼ã‚«ãƒ«ãªå›è»¢ã‚¯ã‚©ãƒ¼ã‚¿ãƒ‹ã‚ªãƒ³ã‚’å–å¾—
#        parent_world_rotation = pbone.parent.matrix.to_quaternion()
#        local_quat = parent_world_rotation.inverted() @ local_quat
#    
#    # ãƒœãƒ¼ãƒ³ã®ãƒ­ãƒ¼ãƒ« (ãƒ­ãƒ¼ã‚«ãƒ«ã®Xè»¸å›è»¢) ã‚’è€ƒæ…®
#    # Blenderã®ãƒœãƒ¼ãƒ³ã¯Yè»¸ãŒãƒœãƒ¼ãƒ³ã®é•·ã•ã«æ²¿ã£ã¦ã„ã‚‹ã“ã¨ãŒå¤šã„ã®ã§ã€ãƒœãƒ¼ãƒ³ã®ãƒ­ãƒ¼ãƒ«ã‚’è£œæ­£
#    # ãƒœãƒ¼ãƒ³ã®ãƒ­ãƒ¼ã‚«ãƒ«Xè»¸ã«æ²¿ã£ãŸå›è»¢ã‚’é©ç”¨
#    # ã“ã®éƒ¨åˆ†ã®å‡¦ç†ã¯ãƒªã‚°ã«ã‚ˆã£ã¦èª¿æ•´ãŒå¿…è¦ã«ãªã‚‹å ´åˆãŒã‚ã‚Šã¾ã™ã€‚
#    
#    # ãƒ­ãƒ¼ã‚«ãƒ«å›è»¢ã‚’è¿”ã™
#    return local_quat

def calculate_rotation(pbone, target_tail_world_loc):
    """
    ç¾åœ¨ã®ãƒœãƒ¼ãƒ³ã®è¦ªã®ãƒˆãƒ©ãƒ³ã‚¹ãƒ•ã‚©ãƒ¼ãƒ ã‚’è€ƒæ…®ã—ã€ãƒ¯ãƒ¼ãƒ«ãƒ‰åº§æ¨™ã®ç›®æ¨™ç‚¹ã«å‘ã‘ã¦ãƒœãƒ¼ãƒ³ã‚’å›è»¢ã•ã›ã‚‹ã‚¯ã‚©ãƒ¼ã‚¿ãƒ‹ã‚ªãƒ³ã‚’è¨ˆç®—ã™ã‚‹
    """
    
    # ã€ä¿®æ­£ç‚¹ã€‘: 'tail_cache' ã¨ 'head_cache' ã‚’ 'tail' ã¨ 'head' (ãƒ¯ãƒ¼ãƒ«ãƒ‰åº§æ¨™) ã«å¤‰æ›´
    # pbone.head/tail ã¯ã€ãƒãƒ¼ã‚ºãƒ¢ãƒ¼ãƒ‰ã®ãƒœãƒ¼ãƒ³ã®ãƒ¯ãƒ¼ãƒ«ãƒ‰åº§æ¨™ã‚’è¿”ã—ã¾ã™ã€‚
    
    pbone_head_world = pbone.head
    pbone_tail_world = pbone.tail 

    # ç¾åœ¨ã®ãƒœãƒ¼ãƒ³ã®ãƒ¯ãƒ¼ãƒ«ãƒ‰åº§æ¨™ã®ãƒ™ã‚¯ãƒˆãƒ« (ãƒœãƒ¼ãƒ³ã®æ–¹å‘)
    # World space bone vector
    current_vec = pbone_tail_world - pbone_head_world
    
    # ç›®æ¨™ã¨ãªã‚‹ãƒ¯ãƒ¼ãƒ«ãƒ‰åº§æ¨™ã®ãƒ™ã‚¯ãƒˆãƒ«
    # ç›®æ¨™ç‚¹ - ãƒœãƒ¼ãƒ³ã®ãƒ˜ãƒƒãƒ‰ã®ãƒ¯ãƒ¼ãƒ«ãƒ‰åº§æ¨™ 
    target_vec = target_tail_world_loc - pbone_head_world
    
    # 2ã¤ã®ãƒ™ã‚¯ãƒˆãƒ«é–“ã®å›è»¢ã‚’è¨ˆç®—
    # rotation_differenceã¯ã€current_vecã‚’target_vecã«å›è»¢ã•ã›ã‚‹ã‚¯ã‚©ãƒ¼ã‚¿ãƒ‹ã‚ªãƒ³ã‚’è¿”ã™
    local_quat = current_vec.rotation_difference(target_vec)
    
    # å›è»¢ã‚’ãƒœãƒ¼ãƒ³ã®ãƒ­ãƒ¼ã‚«ãƒ«ç©ºé–“ã«é©ç”¨ã™ã‚‹ãŸã‚ã«ã€è¦ªã®é€†å›è»¢ã‚’é©ç”¨
    if pbone.parent:
        # è¦ªã®ãƒ¯ãƒ¼ãƒ«ãƒ‰å›è»¢ã®é€†è¡Œåˆ—ã‚’ä½¿ã£ã¦ã€ãƒ­ãƒ¼ã‚«ãƒ«ãªå›è»¢ã‚¯ã‚©ãƒ¼ã‚¿ãƒ‹ã‚ªãƒ³ã‚’å–å¾—
        parent_world_rotation = pbone.parent.matrix.to_quaternion()
        local_quat = parent_world_rotation.inverted() @ local_quat
    
    # ... (ãƒœãƒ¼ãƒ³ã®ãƒ­ãƒ¼ãƒ«ã‚’è€ƒæ…®ã™ã‚‹éƒ¨åˆ†ã¯ãã®ã¾ã¾) ...
    
    # ãƒ­ãƒ¼ã‚«ãƒ«å›è»¢ã‚’è¿”ã™
    return local_quat

# ====================================================================
# 1. ãƒãƒ¼ã‚ºé©ç”¨ãƒ¡ã‚¤ãƒ³é–¢æ•° (å…ƒã®ã‚³ãƒ¼ãƒ‰)
# ====================================================================

def apply_pose_fk_method(armature, keypoints_list):
    """FKï¼ˆãƒ•ã‚©ãƒ¯ãƒ¼ãƒ‰ã‚­ãƒãƒãƒ†ã‚£ã‚¯ã‚¹ï¼‰ãƒ™ãƒ¼ã‚¹ã§ãƒãƒ¼ã‚ºã‚’é©ç”¨ã™ã‚‹ (éšå±¤é †ã§å‡¦ç†)"""
    
    view_layer = bpy.context.view_layer
    
    bpy.ops.object.mode_set(mode='OBJECT')
    view_layer.objects.active = armature
    bpy.ops.object.mode_set(mode='POSE') 
    
    # ãƒãƒ¼ã‚ºãƒœãƒ¼ãƒ³ã®åˆæœŸåŒ–
    for pbone in armature.pose.bones:
        pbone.rotation_mode = 'QUATERNION'
        pbone.rotation_quaternion = (1.0, 0.0, 0.0, 0.0)
        pbone.location = (0.0, 0.0, 0.0)

    # ãƒœãƒ¼ãƒ³éšå±¤ã‚’ã‚½ãƒ¼ãƒˆã—ã¦è¦ªã‹ã‚‰å­ã®é †ã«å‡¦ç†
    # ãƒ’ã‚¨ãƒ©ãƒ«ã‚­ãƒ¼é †ã«ä¸¦ã¹ã‚‹ã“ã¨ã§ã€è¦ªã®ãƒˆãƒ©ãƒ³ã‚¹ãƒ•ã‚©ãƒ¼ãƒ ã‚’å…ˆã«é©ç”¨ã—ã€æ­£ã—ã„åº§æ¨™ã§å­ã®å›è»¢ã‚’è¨ˆç®—ã™ã‚‹
    def get_bone_hierarchy(bone, hierarchy_list):
        # Boneã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆã§ã¯ãªãã€ãã®åå‰ã‚’ãƒªã‚¹ãƒˆã«è¿½åŠ 
        hierarchy_list.append(bone.name) 
        for child in bone.children:
            get_bone_hierarchy(child, hierarchy_list)
        return hierarchy_list

    # ãƒ«ãƒ¼ãƒˆãƒœãƒ¼ãƒ³ã‹ã‚‰ãƒ’ã‚¨ãƒ©ãƒ«ã‚­ãƒ¼ã‚’æ§‹ç¯‰
    sorted_bones = []
    armature_data = armature.data # ã“ã“ã‚’ä¿®æ­£: bpy.data.armatures[armature.data.name] ã‚ˆã‚Šã‚‚å®‰å…¨
    
    if not armature_data.bones:
        print("ã‚¨ãƒ©ãƒ¼: ã‚¢ãƒ¼ãƒãƒãƒ¥ã‚¢ã«ãƒœãƒ¼ãƒ³ãŒã‚ã‚Šã¾ã›ã‚“ã€‚")
        bpy.ops.object.mode_set(mode='OBJECT')
        return

    # å…¨ãƒ«ãƒ¼ãƒˆãƒœãƒ¼ãƒ³ã‹ã‚‰é–‹å§‹
    for bone in armature_data.bones:
        if bone.parent is None:
            get_bone_hierarchy(bone, sorted_bones)
            
    # é‡è¤‡ã‚’å‰Šé™¤ã—ã€é †åºã‚’ç¶­æŒ (è¤‡æ•°ã®ãƒ«ãƒ¼ãƒˆãŒã‚ã‚‹å ´åˆã«å¿…è¦)
    sorted_bones = list(dict.fromkeys(sorted_bones))
            
    # ãƒ’ã‚¨ãƒ©ãƒ«ã‚­ãƒ¼é †ã«ãƒãƒ¼ã‚ºã‚’é©ç”¨
    for bone_name in sorted_bones:
        if bone_name not in armature.pose.bones:
            continue
        
        print(f"ğŸ”„ å‡¦ç†ä¸­ã®ãƒœãƒ¼ãƒ³: {bone_name}")
            
        pbone = armature.pose.bones[bone_name]
        
        # ãƒœãƒ¼ãƒ³ãŒ BONE_INDEX_MAP ã«å«ã¾ã‚Œã¦ã„ã¦ã€ç›®æ¨™åº§æ¨™ãŒã‚ã‚‹å ´åˆã®ã¿å‡¦ç†
        kp_index = BONE_INDEX_MAP.get(bone_name)
        if kp_index is None or kp_index >= len(keypoints_list):
            continue
            
        # å¯¾è±¡ã¨ãªã‚‹ãƒœãƒ¼ãƒ³ãƒ†ãƒ¼ãƒ« (é–¢ç¯€ç‚¹) ã®ç›®æ¨™ãƒ¯ãƒ¼ãƒ«ãƒ‰åº§æ¨™
        target_tail_world_loc = keypoints_list[kp_index]

        if pbone.parent is None: # ãƒ«ãƒ¼ãƒˆãƒœãƒ¼ãƒ³ã®å‡¦ç† (Hipãªã©)
            # ãƒ«ãƒ¼ãƒˆãƒœãƒ¼ãƒ³ï¼ˆHipï¼‰ã®å‡¦ç†: ã‚°ãƒ­ãƒ¼ãƒãƒ«åº§æ¨™ã‚’ãƒ­ãƒ¼ã‚«ãƒ«ä½ç½®ã«å¤‰æ›ã—ã¦è¨­å®š
            # Hip (Index 0) ã®ç›®æ¨™åº§æ¨™ã‚’ã‚¢ãƒ¼ãƒãƒãƒ¥ã‚¢ã®ãƒ­ãƒ¼ã‚«ãƒ«ä½ç½®ã«è¨­å®šã—ã€ãƒãƒ¼ã‚ºãƒœãƒ¼ãƒ³ã®ãƒ­ãƒ¼ã‚«ãƒ«ãƒ˜ãƒƒãƒ‰ä½ç½®ã‚’å¼•ã
            
            # ã€é‡è¦ã€‘Blenderã®ãƒœãƒ¼ãƒ³ã¯Armatureã®åŸç‚¹ã‹ã‚‰ã®ç›¸å¯¾ä½ç½®ã§å®šç¾©ã•ã‚Œã‚‹ãŸã‚ã€
            # pbone.bone.head_local ã¯ãƒœãƒ¼ãƒ³ã® rest position ã®head locationã§ã™
            pbone.location = armature.matrix_world.inverted() @ target_tail_world_loc - pbone.bone.head_local
            
        else: # å­ãƒœãƒ¼ãƒ³ã®å‡¦ç† (å›è»¢è¨ˆç®—)
            try:
                # calculate_rotationã‚’ä½¿ç”¨ã—ã¦ã€ç›®æ¨™ç‚¹ã¸å‘ã‹ã†å›è»¢ã‚¯ã‚©ãƒ¼ã‚¿ãƒ‹ã‚ªãƒ³ã‚’å–å¾—
                local_quat = calculate_rotation(pbone, target_tail_world_loc)
                pbone.rotation_quaternion = local_quat
            except Exception as e:
                print(f"ãƒœãƒ¼ãƒ³ {bone_name} ã®å›è»¢è¨ˆç®—ã‚¨ãƒ©ãƒ¼: {e}")

    bpy.ops.object.mode_set(mode='OBJECT') 
    print("FKãƒ™ãƒ¼ã‚¹ã®ãƒãƒ¼ã‚ºé©ç”¨ãŒå®Œäº†ã—ã¾ã—ãŸã€‚")

# ====================================================================
# 2. ãƒ¡ã‚¤ãƒ³å®Ÿè¡Œé–¢æ•°
# ====================================================================

def run_pose_application():
    # å®Ÿè¡Œå‰ã«ã‚¢ã‚¯ãƒ†ã‚£ãƒ–ãªã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆã®é¸æŠã‚’è§£é™¤
    if bpy.context.view_layer.objects.active:
        bpy.context.view_layer.objects.active.select_set(False)
    
    # --- ã€ãƒ¦ãƒ¼ã‚¶ãƒ¼è¨­å®šã‚¨ãƒªã‚¢ã€‘ ---
    
    # 1. èª­ã¿è¾¼ã‚€NPZãƒ•ã‚¡ã‚¤ãƒ«ã®ãƒ‘ã‚¹ã‚’æŒ‡å®šã—ã¦ãã ã•ã„
    # ä¾‹: "C:\\Users\\User\\Desktop\\my_keypoints.npz"
    NPZ_FILEPATH = "C:/Users/a24k0/R6_blender/scripts/keypoints.npz" # "//" ã¯Blenderãƒ•ã‚¡ã‚¤ãƒ«ã‹ã‚‰ã®ç›¸å¯¾ãƒ‘ã‚¹
    
    # 2. Blenderã‚·ãƒ¼ãƒ³å†…ã®å¯¾è±¡ã‚¢ãƒ¼ãƒãƒãƒ¥ã‚¢åã‚’è¨­å®šã—ã¦ãã ã•ã„
    ARMATURE_NAME = "Armature" 
    
    # --- å‡¦ç†ã®é–‹å§‹ ---
    
    try:
        armature = bpy.data.objects[ARMATURE_NAME]
    except KeyError:
        print(f"âŒ ã‚¨ãƒ©ãƒ¼: ã‚¢ãƒ¼ãƒãƒãƒ¥ã‚¢ '{ARMATURE_NAME}' ãŒã‚·ãƒ¼ãƒ³ã«è¦‹ã¤ã‹ã‚Šã¾ã›ã‚“ã€‚")
        return
        
    # 1. NPZãƒ•ã‚¡ã‚¤ãƒ«ã®èª­ã¿è¾¼ã¿
    keypoints_list = load_keypoints_data_from_npz(bpy.path.abspath(NPZ_FILEPATH))
    
    if keypoints_list is None:
        print("å‡¦ç†ã‚’ä¸­æ–­ã—ã¾ã™ã€‚")
        return
        
    # 2. ãƒãƒ¼ã‚ºã®é©ç”¨
    apply_pose_fk_method(armature, keypoints_list)

# å®Ÿè¡Œ
if __name__ == "__main__":
    run_pose_application()
```
